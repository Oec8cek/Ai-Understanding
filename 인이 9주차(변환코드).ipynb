기존 데이터 생성 코드

```python
import nltk
from nltk.corpus import *
import random
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import *
```

```python
labeled_names = ([(name, 0) for name in names.words('male.txt')]
                + [(name, 1) for name in names.words('female.txt')])

random.shuffle(labeled_names)

def gender_features (word) :
    return word[-1]
alphabet = [' ', 'a', 'b', 'c', 'd', 'e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
featuresets = [(alphabet.index(gender_features (n)), gender) for (n,gender) in labeled_names]

featuresets
```

pandas를 이용해 data를 정리

```python
dataset = pd.DataFrame(featuresets, columns = ['last_letter', 'gender'])
dataset
```

# train_test_split 함수 (arrays, test_size, train_size, random_state, shuffle, stratify)

arrays : 분할시킬 데이터를 입력해야 한다. numpy array, pandas df, list 형태 모두 가능하다

test_size : 테스트 할 데이터셋의 비율(float)이나 개수(int)를 입력해야 한다. (default = 0.25)

train_size : 학습 시킬 데이터셋의 비율(float)이나 개수(int) (default = test_size의 나머지)

shuffle : 셔플여부 (default = True)

# 함수가 return 해주는 값

X_train, X_test, Y_train, Y_test : arrays에 데이터와 레이블을 둘 다 넣었을 경우의 반환이며, 데이터와 레이블의 순서쌍은 유지된다.

X_train, X_test : arrays에 레이블 없이 데이터만 넣었을 경우의 반환

함수가 어려워서 도움을 받기 위해 적어놓음. [출처] [Python] sklearn의 train_test_split() 사용법|작성자 Paris Lee

```python
train_set, test_set = train_test_split(dataset, test_size = 0.06)

train_set
```

# Fit Naive Bayes classifier according to X, y.

fit(X, y[, sample_weight])

```python
# 2d array 여야 한다고 오류가 떠서 추가한 코드들 
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer()
train_set_letter = np.array(train_set['last_letter'])
train_set_letter = train_set_letter.reshape(-1, 1)
imp.fit_transform(train_set_letter)

test_set_letter = np.array(test_set['last_letter'])
test_set_letter = test_set_letter.reshape(-1, 1)
imp.fit_transform(test_set_letter)
```

```python
# train 시키기 
clf = GaussianNB()
clf.fit(train_set_letter, train_set['gender'])

# 정확도 측정하기 
clf.score(test_set_letter, test_set['gender'])
```
