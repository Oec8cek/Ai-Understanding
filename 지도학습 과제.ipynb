```python
import sys, fileinput, re
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
```

```python
if raw.strip() != "":
    raw = re.sub(r'([a-z])\.([A-Z])', r'\1. \2', raw.strip())
    sentences = sent_tokenize(raw.strip())
    for s in sentences:
      if s != "":
          result = sys.stdout.write(s+"\n")

result
```

```python
# 형태소 분석 

# 1. 어간 추출
# 2. 원형 복원
# 3. 품사 부착
```

```python
# 어간 추출
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
st1 = PorterStemmer()
st2 =  LancasterStemmer()
```

```python
words = word_tokenize(raw.strip())
```

```python
print("Porter Stemmer :", [st1.stem(w) for w in words])
print("Lancaster Stemmer :", [st2.stem(w) for w in words])
```

```python
# 원형 복원
# 같은 의미를 가지는 여러 단어를 사전형으로 통일하는 작업
# 품사(part of speech)를 지정하는 경우 좀 더 정확한 원형을 찾기 가능
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')

WNL = WordNetLemmatizer()
[WNL.lemmatize(w, pos='v') for w in words]
```
